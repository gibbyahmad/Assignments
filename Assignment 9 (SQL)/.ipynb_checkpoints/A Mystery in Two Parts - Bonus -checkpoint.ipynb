```python
#Import Dependencies
from sqlalchemy import create_engine

import pandas as pd

import matplotlib.pyplot as plt

import numpy as np

```

```python
#Create Engine to pass in Postgres

engine = create_engine('postgresql://postgres:teeToo101@localhost:5432/SQL_Homework')
connection = engine.connect()
```

```python
#Query all Records in the Salaries Table
salaries_data = pd.read_sql("SELECT * FROM salaries", connection)
salaries_data.head()
```

```python
#Query all records in the titles table
titles_data = pd.read_sql("SELECT * FROM titles", connection)
titles_data.head()
```

```python
#Merge Columns / combine the data
combined_data = pd.merge(salaries_data, titles_data, on="emp_no", how="inner")
combined_data.head()
```

```python
# Groupby Title to show mean salary
grouped_df = combined_data.groupby("title").mean()
grouped_df
```

```python
#drop emp_no from grouped_df to show only title and salary
revised_grouped_df = grouped_df.drop(columns = "emp_no")
revised_grouped_df
```

```python
#Reset Index
revised_grouped_df = revised_grouped_df.reset_index()
revised_grouped_df
```

```python
#Plot Graph
#x-axis, y-axis, tick location
x_axis = revised_grouped_df["title"]
ticks = np.arange(len(x_axis))
y_axis = revised_grouped_df["salary"]

#create bar chart
plt.bar(x_axis, y_axis, align="center", alpha=0.5, color=["y", "c", "b", "m", "g", "r", "k"])

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="vertical")

# Set Labels & Title
plt.ylabel("Salaries ($)")
plt.xlabel("Employee Titles")
plt.title("Average Employee Earnings by Title")

#Save the figure
plt.savefig("./average_earnings_by_title.png")

#show plot
plt.show()
```
